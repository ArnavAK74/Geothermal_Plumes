```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to filter dataset for a specific time range and display the results
def filter_and_save_time_range(file_path, output_path, start_time, end_time):
    # Load the dataset
    df = pd.read_csv(file_path, parse_dates=["Time"])

    # Compute Temperature Difference (Temp_Diff)
    df["Temp_Diff"] = df["Temperature"].diff()

    # Drop NaN values (first row will have NaN)
    df = df.dropna()

    # Filter for the specified time range
    filtered_data = df[(df['Time'] >= start_time) & (df['Time'] <= end_time)]

    # Save the filtered dataset
    filtered_data.to_csv(output_path, index=False)
    print(f"Filtered dataset saved to {output_path}")

    # Plot the original and filtered data
    plt.figure(figsize=(14, 6), dpi=100)
    plt.plot(df['Time'], df['Temperature'], label='Original Data', color='blue', linewidth=0.5, alpha=0.7)
    plt.plot(filtered_data['Time'], filtered_data['Temperature'], label='Filtered Data', color='red', linewidth=1.5)
    plt.title(f"Original vs Filtered Data for {file_path.split('/')[-1]}")
    plt.xlabel("Time")
    plt.ylabel("Temperature (°C)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# List of thermistor datasets with file paths and output paths
datasets = [
    {"file": "/content/drive/MyDrive/Research/0411/T1_0411.csv", "output": "/content/drive/MyDrive/Research/Refined_0411/Refined_T1_0411.csv"},
    {"file": "/content/drive/MyDrive/Research/0411/T2_0411.csv", "output": "/content/drive/MyDrive/Research/Refined_0411/Refined_T2_0411.csv"},
    {"file": "/content/drive/MyDrive/Research/0411/T3_0411.csv", "output": "/content/drive/MyDrive/Research/Refined_0411/Refined_T3_0411.csv"},
    {"file": "/content/drive/MyDrive/Research/0411/T4_0411.csv", "output": "/content/drive/MyDrive/Research/Refined_0411/Refined_T4_0411.csv"},
    {"file": "/content/drive/MyDrive/Research/0411/T7_0411.csv", "output": "/content/drive/MyDrive/Research/Refined_0411/Refined_T7_0411.csv"},
    {"file": "/content/drive/MyDrive/Research/0411/T12_0411.csv", "output": "/content/drive/MyDrive/Research/Refined_0411/Refined_T12_0411.csv"},
    {"file": "/content/drive/MyDrive/Research/0411/T13_0411.csv", "output": "/content/drive/MyDrive/Research/Refined_0411/Refined_T13_0411.csv"},
    {"file": "/content/drive/MyDrive/Research/0411/T16_0411.csv", "output": "/content/drive/MyDrive/Research/Refined_0411/Refined_T16_0411.csv"},
    {"file": "/content/drive/MyDrive/Research/0411/T17_0411.csv", "output": "/content/drive/MyDrive/Research/Refined_0411/Refined_T17_0411.csv"},
]

# Specify the time range for filtering
start_time = "2023-04-11 17:09:30"
end_time = "2023-04-11 17:50:00"

# Run the filtering and display the plots for all datasets
for dataset in datasets:
    filter_and_save_time_range(
        file_path=dataset["file"],
        output_path=dataset["output"],
        start_time=start_time,
        end_time=end_time
    )
```

```python
import pandas as pd

# List of thermistor dataset file paths
file_paths = [
    "/content/drive/MyDrive/Research/Refined_0411/Refined_T1_0411.csv",
    "/content/drive/MyDrive/Research/Refined_0411/Refined_T2_0411.csv",
    "/content/drive/MyDrive/Research/Refined_0411/Refined_T3_0411.csv",
    "/content/drive/MyDrive/Research/Refined_0411/Refined_T4_0411.csv",
    "/content/drive/MyDrive/Research/Refined_0411/Refined_T7_0411.csv",
    "/content/drive/MyDrive/Research/Refined_0411/Refined_T12_0411.csv",
    "/content/drive/MyDrive/Research/Refined_0411/Refined_T13_0411.csv",
    "/content/drive/MyDrive/Research/Refined_0411/Refined_T16_0411.csv",
    "/content/drive/MyDrive/Research/Refined_0411/Refined_T17_0411.csv"
]

# Create an empty list to store DataFrames
df_list = []

# Load and merge datasets
for file_path in file_paths:
    thermistor_id = file_path.split("/")[-1].split("_")[1]  # Extract thermistor name (T1, T2, etc.)
    df = pd.read_csv(file_path, parse_dates=["Time"])
    df["Thermistor_ID"] = thermistor_id  # Assign Thermistor ID
    df_list.append(df)

# Concatenate all datasets
merged_df = pd.concat(df_list, ignore_index=True)

# Save merged dataset
merged_df.to_csv("/content/drive/MyDrive/Research/Refined_0411/Merged_Thermistor_Data.csv", index=False)

print("Merged dataset saved successfully!")
```
